```python
from matplotlib import pyplot as plt
from math import log
import numpy as np

def scaled_speedup(n,ps):
    communication_cost=20
    calculation_cost=1
    speedup=[]
    for p in ps:
        totalwork=n*p
        localwork=totalwork/p*calculation_cost
        interwork=log(p)*(communication_cost+calculation_cost)
        speedup.append(totalwork/(localwork+interwork))
    return speedup

ps=np.array([1,2**2,2**4,2**5,2**8])
standard=[1 for i in ps]

plt.plot(np.log2(ps), standard,label="standard")
plt.plot(np.log2(ps), scaled_speedup(256,ps),label="parallel")
plt.xlabel("2^x threads")
plt.ylabel("scaled speedup")
plt.legend(loc='upper left')
plt.show()
```
